# Challenge 2 - Hypothesis Tests

In this section of the lab, we will test whether the hourly wage of all hourly workers is significantly different from $30/hr. Import the correct one sample test function from scipy and perform the hypothesis test for a 95% two sided confidence interval.

```python
# Hypothesis Test 1: Test if the hourly wage of all hourly workers is significantly different from $30/hr
from scipy.stats import ttest_1samp

hourly_wages = salaries_df[salaries_df["Salary or Hourly"] == "Hourly"]["Hourly Rate"].dropna()
hourly_wages

# Perform one-sample t-test
h0 = 30  # Null hypothesis-Hourly wage is $30/hr
alpha = 0.05  # Significance level

stat, pval = ttest_1samp(hourly_wages,h0)

print(stat)
print(pval)

# Check if the null hypothesis can be rejected
if pval < alpha:
    print("Reject the null hypothesis: The mean hourly wage of all hourly workers is different from $30/hr.")
else:
    print("Fail to reject the null hypothesis.")
```

We are also curious about salaries in the police force. The chief of police in Chicago claimed in a press briefing that salaries this year are higher than last year's mean of $86000/year a year for all salaried employees. Test this one sided hypothesis using a 95% confidence interval.

Hint: A one tailed test has a p-value that is half of the two tailed p-value. If our Null hypothesis is less than or equal to, then to reject, the test statistic must be positive (rejects the <) and the p-value below 0.05 (rejects the =).

```python
# Salary  in the police force

police_salaries = salaries_df[(salaries_df["Department"] == "POLICE") &
                                               (salaries_df["Salary or Hourly"] == "Salary")]["Annual Salary"].dropna()

# Perform one-sample t-test

h0 = 86000  # Null hypothesis: Salaries this year are not higher than last year's mean of $86000/year
alpha = 0.05  # Significance level
stat, pval = ttest_1samp(police_salaries, h0)

# One-tailed test

print('stat is  ', stat)
print('pvalue for the two-tailed test is ', pval/2)

# Check if the null hypothesis can be rejected
if stat > 0 and pval < alpha:
    print("Reject the null hypothesis: Salaries this year are higher than last year's mean of $86000/year.")
else:
    print("Fail to reject the null hypothesis.")
    

#if the t-statistic is positive, it indicates that the sample mean is higher 
#than the hypothesized value ($86000/year), providing evidence to reject the null hypothesis in favor of the alternative hypothesis.
```

Using the `crosstab` function, find the department that has the most hourly workers. 

```python
# Crosstab to find department with the most hourly workers
hourly_department = pd.crosstab(salaries_df["Department"], salaries_df["Salary or Hourly"]).sort_values(by="Hourly", ascending=False).index[0]
print("Department with the most hourly workers:", hourly_department)


```

The workers from the department with the most hourly workers have complained that their hourly wage is less than $35/hour. Using a one sample t-test, test this one-sided hypothesis at the 95% confidence level.

```python
# Hourly wages in the department with the most hourly workers
hourly_wages_most_hourly = salaries_df[(salaries_df["Department"] == hourly_department) & (salaries_df["Salary or Hourly"] == "Hourly")]["Hourly Rate"].dropna()

# Perform one-sample t-test
h0 = 35  # Null hypothesis: Hourly wage is $35/hr
alpha = 0.05  # Significance level
stat, pval = ttest_1samp(hourly_wages_most_hourly, h0)

# One-tailed test
pval /= 2

# Check if the null hypothesis can be rejected
if stat < 0 and pval < alpha:
    print("Reject the null hypothesis: Hourly wage in the department with the most hourly workers is less than $35/hr.")
else:
    print("Fail to reject the null hypothesis.")

#we are interested in detecting a negative difference. Therefore, if the t-statistic is negative, it indicates
#that the sample mean is lower than the hypothesized 
#value ($35/hr), providing evidence to reject the null hypothesis in favor of the alternative hypothesis.
```

# Challenge 3: To practice - Constructing Confidence Intervals

While testing our hypothesis is a great way to gather empirical evidence for accepting or rejecting the hypothesis, another way to gather evidence is by creating a confidence interval. A confidence interval gives us information about the true mean of the population. So for a 95% confidence interval, we are 95% sure that the mean of the population is within the confidence interval. 
).

To read more about confidence intervals, click [here](https://en.wikipedia.org/wiki/Confidence_interval).


In the cell below, we will construct a 95% confidence interval for the mean hourly wage of all hourly workers. 

The confidence interval is computed in SciPy using the `t.interval` function. You can read more about this function [here](https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.t.html).

To compute the confidence interval of the hourly wage, use the 0.95 for the confidence level, number of rows - 1 for degrees of freedom, the mean of the sample for the location parameter and the standard error for the scale. The standard error can be computed using [this](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.sem.html) function in SciPy.

```python
from scipy.stats import t
import scipy.stats
import numpy as np

# Calculate sample mean and standard error
sample_mean = hourly_wages.mean()
sample_std = hourly_wages.std()
sample_standard_error = scipy.stats.sem(hourly_wages)

# Compute 95% confidence interval
confidence_level = 0.95
degrees_freedom = len(hourly_wages) - 1
confidence_interval = t.interval(confidence_level, 
                                             degrees_freedom, 
                                             sample_mean, 
                                             sample_standard_error)

print("95% Confidence Interval for mean hourly wage of all hourly workers:", confidence_interval)

```

Now construct the 95% confidence interval for all salaried employeed in the police in the cell below.

```python
# Calculate sample mean and standard error for police salaries
police_sample_mean = police_salaries.mean()
police_sample_std = police_salaries.std()
degrees_freedom = len(police_salaries)-1
police_standard_error = scipy.stats.sem(police_salaries)

# Compute 95% confidence interval for police salaries
confidence_level = 0.95
police_confidence_interval = t.interval(confidence_level, degrees_freedom,police_sample_mean, 
                                             police_standard_error)

print("95% Confidence Interval for mean annual salary of salaried police employees:", police_confidence_interval)
```
